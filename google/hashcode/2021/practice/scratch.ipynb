```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.graph_objects as go
import plotly.express as px
from tqdm import tqdm

import os
import time

T = 101**2 * 0.25
# f(0) = T
# f(K) -> 0
f = lambda k : T * 1 / np.log(k*T+np.exp(1))

K = 1e3

fig = go.Figure()
fig.add_trace(go.Scatter(x=np.arange(K), y=np.vectorize(f)(np.arange(K))))
fig.show()
```
